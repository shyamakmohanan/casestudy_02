# Who scores the most on average for math, reading and writing
# Based on Gender

```python
df['Average'] = df.mean(axis=1)
df2 = df[df.gender=='female']
df3 = df[df.gender=='male']
```

```python
print('gender - MALE')
Male_MathAverage = df3['math score'].mean()
print('Average math score is',round(Male_MathAverage))
Male_ReadAverage = df3['reading score'].mean()
print('Average reading score is',round(Male_ReadAverage))
Male_WriteAverage = df3['writing score'].mean()
print('Average writing score is',round(Male_WriteAverage))
```

```python
print('gender - FEMALE')
Female_MathAverage = df2['math score'].mean()
print('Average math score is', round(Female_MathAverage))
Female_ReadAverage = df2['reading score'].mean()
print('Average read score is', round(Female_ReadAverage))
Female_WriteAverage = df2['writing score'].mean()
print('Average write score is', round(Female_WriteAverage))
```

Average male math score is 69 compared to female score of 63. Male tops the most in math score.
Average male reading score is 66 compared to female score of 73. Female tops the most in reading score.
Average male writing score is 63 compared to female score of 72. Female tops the most in writing score.

# Who scores the most on average for math, reading and writing
# Based on Test Preparation Course

```python
df4 = df['test preparation course']=='none'
df4 = df[df4]
print("Test preparation course - none")
print('Average math score is', round(df4['math score'].mean()))
print('Average reading score is', round(df4['reading score'].mean()))
print('Average writing score is', round(df4['writing score'].mean()))
```

```python
df5 = df['test preparation course']=='completed'
df5 = df[df5]
print("Test preparation course - completed")
print('Average math score is',round(df5['math score'].mean()))
print('Average reading score is', round(df5['reading score'].mean()))
print('Average writing score is', round(df5['writing score'].mean()))
```

From above outputs it is clear that students whom completed test preparation course scores tops in all the scores when compared to those whom not completed.

# Scoring variation for math, reading and writing based on 'Gender'

```python
print('gender - MALE')
Male_MathVariation = df3['math score'].var()
print('Scoring variation for math score is',round(Male_MathVariation))
Male_ReadVariation = df3['reading score'].var()
print('Scoring variation for reading score is',round(Male_ReadVariation))
Male_WriteVariation = df3['writing score'].var()
print('Scoring variation for writing score is',round(Male_WriteVariation))
```

```python
print('gender - FEMALE')
Female_MathVariation = df2['math score'].var()
print('Scoring variation for math score is', round(Female_MathVariation))
Female_ReadVariation = df2['reading score'].var()
print('Scoring variation for read score is', round(Female_ReadVariation))
Female_WriteVariation = df2['writing score'].var()
print('Scoring variation for write score is', round(Female_WriteVariation))
```

From the above output it is clear that female score variation is high compared to males in math, reading and writing.
